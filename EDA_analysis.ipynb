# Data Analysis & Visualizations

### Which states have the highest and lowest EV sales?

```python
state_sales = df.groupby('State')['EV_Sales_Quantity'].sum().reset_index() # Group the data by state and sum the EV sales for each state
state_sales = state_sales.sort_values(by = ['EV_Sales_Quantity'], ascending = False)

plt.figure(figsize= (15,6))
a = sns.barplot(x = 'State', y = 'EV_Sales_Quantity', data = state_sales)
plt.xlabel('State')
plt.ylabel('EV sales')
plt.xticks(rotation = 90)
plt.title('Total Sales')

for bars in a.containers: # Annotate each bar with the respective count
    a.bar_label(bars, label_type ='edge', rotation =90, padding =3)

plt.show()
```

### What is the distribution of EV sales across different vehicle types?

```python
vehicle_sale = df.groupby('Vehicle_Type')['EV_Sales_Quantity'].sum().reset_index() # Grouping 'EV_Sales_Quantity' by 'Vehicle_Type'
vehicle_sale = vehicle_sale.sort_values(by = ['EV_Sales_Quantity'], ascending = False)

plt.figure(figsize =(18,6))
a = sns.barplot(x='Vehicle_Type', y='EV_Sales_Quantity', data = vehicle_sale)
plt.title('EV sales across different Vehicle types')
plt.xlabel('Vehicle Type')
plt.xticks(rotation = 60)
plt.ylabel('Total Sales')

for bars in a.containers: # Annotate each bar with the respective count
    a.bar_label(bars)
plt.show()
```

### Which vehicle category is the most popular?

```python
vehicle_category = df.groupby('Vehicle_Category')['EV_Sales_Quantity'].sum().reset_index() # Grouping Vehicle category and Total Quantity Sold
vehicle_category = vehicle_category.sort_values(by = 'EV_Sales_Quantity', ascending= False)

plt.figure(figsize=(8,4))
a = sns.barplot( x ='Vehicle_Category', y = 'EV_Sales_Quantity', data = vehicle_category)
plt.title('Most sold category of Vehicle')
plt.xlabel('Vehicle Category')
plt.ylabel('Sales Quantity')

for bars in a.containers:
    a.bar_label(bars)
plt.show()
```

### How does the adoption of different vehicle types vary across states?

```python
state_vehicle_sales = df.pivot_table(values = 'EV_Sales_Quantity', index = 'State', columns = 'Vehicle_Type', aggfunc= 'sum', fill_value= 0)

plt.figure(figsize = (14,10))
sns.heatmap(state_vehicle_sales, fmt ='g', cmap='Reds', linecolor='k', linewidths= 0.5 )
plt.title('EV sales by State and Vehicle Type')
plt.xticks(rotation =90)
plt.show()
```

### How have EV sales evolved over different years?

```python
sorted_data = df[['Year', 'EV_Sales_Quantity']] # Creating temporary data set having 'Year' & 'EV_Sales_Quantity' only

plt.figure(figsize = (12,6))
sns.lineplot(x  ='Year', y = 'EV_Sales_Quantity', data = sorted_data, marker='o')
plt.title('EV Sales Quantity by Year (Descending Order)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.savefig('EV Sales Quantity by Year (Descending Order)')
plt.show()
```
